* 'Address', 'Cross Street': This is an approximate location

#### **Q7:** A function to pull out longitude

Now, let's look at the `Location` field.  Each location field value is formatted as a string similar to this:
```
(34.0416, -118.2896)
```

We'd like to convert that to a numeric latitude and longitude.  Here's a function that takes a couple of steps to get the latitude out (that's the first number, e.g. `34.0416` in this case)

* It first gets rid of the `(` and `)` characters using `replace`.
* It then uses "split" to turn the string in a list of strings, with the part before the comma in `[0]` and the part after the comma in `[1]`
* It then converts the string value to a `float` and returns it.

````python
def location_to_lat(location):
    withoutParens = location.replace("(","").replace(")","")
    latLongList = withoutParens.split(",")
    lat = float(latLongList[0])
    return lat

location_to_lat("(34.0416, -118.2896)")
````
